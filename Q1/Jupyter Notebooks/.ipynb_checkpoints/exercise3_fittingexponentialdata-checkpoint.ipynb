Next we'll quickly load the data and make a quick plot out of it, this is all familiar territory now.

```python
t, A, dA = np.loadtxt(data_filename, unpack=True, skiprows=1)
```

```python
fig, ax = plt.subplots(figsize = (10,8))
ax.errorbar(t, A, dA)
print(A)
```

Okay, it isn't clear what is going on at a glance.  Let's see if a linear fit is any good. The two cells below should look familiar, they're copied from the last notebook.

```python
def linear(p, xvar):
    return p[0] + p[1]*xvar

def residual(p,func, xvar, yvar, err):
    return (func(p, xvar) - yvar)/err
```

```python
# The code below defines our data fitting function.
# Inputs are:
# initial guess for parameters p0
# the function we're fitting to
# the x,y, and dy variables
# tmi can be set to 1 or 2 if more intermediate data is needed

def data_fit(p0,func,xvar, yvar, err,tmi=0):
    try:
        fit = optimize.least_squares(residual, p0, args=(func,xvar, yvar, err),verbose=tmi)
    except Exception as error:
        print("Something has gone wrong:",error)
        return p0, np.zeros_like(p0), -1, -1
    pf = fit['x']

    print()

    try:
        cov = np.linalg.inv(fit['jac'].T.dot(fit['jac']))          
        # This computes a covariance matrix by finding the inverse of the Jacobian times its transpose
        # We need this to find the uncertainty in our fit parameters
    except:
        # If the fit failed, print the reason
        print('Fit did not converge')
        print('Result is likely a local minimum')
        print('Try changing initial values')
        print('Status code:', fit['status'])
        print(fit['message'])
        return pf,np.zeros_like(pf), -1, -1
            #You'll be able to plot with this, but it will not be a good fit.

    chisq = sum(residual(pf,func,xvar, yvar, err) **2)
    dof = len(xvar) - len(pf)
    red_chisq = chisq/dof
    pferr = np.sqrt(np.diagonal(cov)) # finds the uncertainty in fit parameters by squaring diagonal elements of the covariance matrix
    print('Converged with chi-squared {:.2f}'.format(chisq))
    print('Number of degrees of freedom, dof = {:.2f}'.format(dof))
    print('Reduced chi-squared {:.2f}'.format(red_chisq))
    print()
    Columns = ["Parameter #","Initial guess values:", "Best fit values:", "Uncertainties in the best fit values:"]
    print('{:<11}'.format(Columns[0]),'|','{:<24}'.format(Columns[1]),"|",'{:<24}'.format(Columns[2]),"|",'{:<24}'.format(Columns[3]))
    for num in range(len(pf)):
        print('{:<11}'.format(num),'|','{:<24.3e}'.format(p0[num]),'|','{:<24.3e}'.format(pf[num]),'|','{:<24.3e}'.format(pferr[num]))
    return pf, pferr, chisq, dof
```

```python
print("Linear Fit")
guess = [10., 10.]
pf, pferr, chisq, dof = data_fit(guess,linear, t, A, dA)
print(pf, pferr, dof)
```

```python
fig, ax = plt.subplots(figsize = (10,8))
ax.errorbar(t, A, dA, fmt='k.', label = "Data")
t_cont = np.linspace(min(t), max(t), 200) 
  # This makes it so we get a smooth plot rather than only the function evaluated exactly at our data points
  # It isn't strictly needed for a straight line, but it will make other functions look much nicer
ax.plot(t_cont, linear(pf, t_cont), label = "Linear fit")
ax.legend()
```

Oh. The reduced $\chi^2$ value is 26.9; that's a good indication that a linear function isn't a very good model for our data. Remember that this is the best match our algorithm could find for **any** linear function. All linear functions with different slopes and intercepts would be even worse.

# Part 2: Exponential Functions
If all went well, you probably saw that your quadratic function makes for a much better fit for this data than a straight line did.  However, there's a problem with calling it a day now: this is from an exponential decay!  So, even though a quadratic function sort of fits, if we want to know the lifetime $\tau$ of the quantity in question we'll need to fit to an exponential instead, of the form $Ae^{-\frac{x}{\tau}}$.

```python
def expfunc(p, x):
    return p[0]*np.exp(-x/p[1])
```

With that defined, let's try and fit our data with this function as usual.

```python
guess_exp = [10, 1]
pf_exp, pferr_exp, chisq_exp, dof_exp = data_fit(guess_exp, expfunc, t, A, dA) 

fig, ax = plt.subplots(figsize = (10,8))
ax.errorbar(t, A, dA, fmt='k.', label = "Data")
t_cont = np.linspace(min(t), max(t), 200)
ax.plot(t_cont, linear(pf, t_cont), label = "Linear fit: $\chi^2/N = {:0.2f}$".format(chisq/dof))
ax.plot(t_cont, expfunc(pf_exp, t_cont), label = "Exponential fit: $\chi^2/N = {:0.2f}$".format(chisq_exp/dof_exp))
ax.legend()
```

That doesn't look right at all, what happened?

Notice that the data_fit function gave a different message than before: the fit didn't converge.

For clues, let's look at what our function looked like with our initial parameters.


```python
guess_exp = [10, 1]
pf_exp, pferr_exp, chisq_exp, dof_exp = data_fit(guess_exp, expfunc, t, A, dA) 

fig, ax = plt.subplots(figsize = (10,8))
ax.errorbar(t, A, dA, fmt='k.', label = "Data")
t_cont = np.linspace(min(t), max(t), 200)
ax.plot(t_cont, linear(pf, t_cont), label = "Linear fit: $\chi^2/N = {:0.2f}$".format(chisq/dof))
ax.plot(t_cont, expfunc(pf_exp, t_cont), label = "Exponential fit: $\chi^2/N = {:0.2f}$".format(chisq_exp/dof_exp))
ax.plot(t_cont, expfunc(guess_exp, t_cont), label = "Exponential fit guess")
ax.legend()
```

This might make a little more sense now: our initial guess (shown in green) resulted in a function where almost all of the values evaluated to zero.  As a result, the residuals for all but the first point were calculated to be enormous.  No small change in the initial parameters could fix this, so instead the algorithm converged to a function that minimized the residual at the first point: $f(x) = 125.2e^{-x}$

In the cell below, modify the initial guess to get a function that fits the data well.

```python
guess_exp = [100, 300]
pf_exp, pferr_exp, chisq_exp, dof_exp = data_fit(guess_exp, expfunc, t, A, dA) 

fig, ax = plt.subplots(figsize = (10,8))
ax.errorbar(t, A, dA, fmt='k.', label = "Data")
t_cont = np.linspace(min(t), max(t), 200)
ax.plot(t_cont, linear(pf,t_cont), label = "Linear fit: $\chi^2/N = {:0.2f}$".format(chisq/dof))
ax.plot(t_cont, expfunc(pf_exp, t_cont), label = "Exponential fit: $\chi^2/N = {:0.2f}$".format(chisq_exp/dof_exp))
ax.plot(t_cont, expfunc(guess_exp, t_cont), label = "Exponential fit guess")
ax.legend()
```

There's one more thing we can do to better model the physics of this system.  When we're monitoring the radioactive decay of something, there is often a background term due to noise in the detector itself.  This background doesn't change with time, so it is reflected by adding a constant background term to our function: $f(x) = A e^{-\frac{x}{\ell}} + C$

Let's try it in the cell below.

```python
def expfunc_bg(p, x):
    return (p[0]*np.exp(-x/p[1])) + p[2]

#def expfunc_bg(p, x):
#    return expfunc([p[0],p[1]],x) + p[2]
# This is an alternate way you could define the background function by re-using the exponential function we already created.

guess_exp_bg = [100, 500, -10]
guess_exp = [100, 500]
pf_exp_bg, pferr_exp_bg, chisq_exp_bg, dof_exp_bg = data_fit(guess_exp_bg, expfunc_bg, t, A, dA) 
pf_exp, pferr_exp, chisq_exp, dof_exp = data_fit(guess_exp, expfunc, t, A, dA) 

fig, ax = plt.subplots(figsize = (10,8))
ax.errorbar(t, A, dA, fmt='k.', label = "Data")
t_cont = np.linspace(min(t), max(t), 200)
ax.plot(t_cont, linear(pf, t_cont), label = "Linear fit: $\chi^2/N = {:0.2f}$".format(chisq/dof))
ax.plot(t_cont, expfunc(pf_exp, t_cont), label = "Exponential fit: $\chi^2/N = {:0.2f}$".format(chisq_exp/dof_exp))
ax.plot(t_cont, expfunc_bg(pf_exp_bg, t_cont), label = "Exponential fit with background: $\chi^2/N = {:0.2f}$".format(chisq_exp_bg/dof_exp_bg))
ax.legend()
```

The most notable difference here is in the last data point: the fit of an exponential with a background term is a much better fit than the one without it.

This difference is also reflected in the reduced $\chi^2$ values: 1.45 for the exponential fit and 1.16 for the one with a background term.

In general, if you have multiple possible models for your data, the one which fits with the lowest reduced $\chi^2$ value is better.  

_However_, just because a model produces a low reduced $\chi^2 \approx 1.6$ value doesn't mean it is correct.  For instance, a quadratic function produced a pretty good fit for our data, but if you tried to predict what would happen at later times you'd get a disturbingly unphysical result.

# Part 5

Assuming that went well, its time to address the uncertainty in the net count rates, `dR`.  Since $R = \frac{G}{t}$, when we calculate the uncertainty we get $\frac{dR}{R} \approx \frac{dG}{G}$. (We're ignoring the uncertainty in time here, because it contributes negligibly.)

Now the question is how do we get the uncertainty in the net counts, $dG$?

Since $dG$ is a number acquired from counting statistics, its uncertainty is just $dG = \sqrt{G}$

Putting things together with our expression for $dR$, we get $dR = \frac{R \sqrt{dG^2}}{G}= \frac{R\sqrt{G}}{G}$
